## Univariate Analysis

#### Numerical Variables

```python
def numeric_univariate(column):
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # Boxplot
    sns.boxplot(ax=axes[0], data=df, x=column)
    axes[0].set_title(f"{column} (Boxplot)")

    # Distribution
    sns.boxplot(ax=axes[1], data=df, x=column)
    axes[1].set_title(f"{column} (KDE)")
    
    
    plt.show()
numeric_univariate("Total night minutes")
```
