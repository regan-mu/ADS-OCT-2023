## CONDITIONAL STATEMENTS.
- Programming often involves examining a set of conditions and deciding which action to take based on those conditions.
- Pythonâ€™s if statement allows you to examine the current state of a program and respond appropriately to that state.
- Several different kinds of conditional statements exist, and your choice of which to use depends on the number of conditions you need to test.

##### Simple IF statement.

* This is the simplest kind of conditional statement has one test and one action: <br>

        if conditional_test:
            do something
            
* If the conditional test evaluates to True the code following the statement is executed.

```python
age = 19
if age >= 18:
    print("You are old enough to vote!")
```

##### if-else Statements
- Quite often you want to take one action when the conditional test evaluates to True and a different action in all other cases.
- Pythonâ€™s **if-else** block makes this possible.
- An if-else block is similar to a simple if statement, but the else statement allows you to define an action or set of actions that are executed when the conditional test fails.

        if conditional_test:
            do something
        else:
            do something else

```python
age = 17
if age >= 18:
    print("You are old enough to vote!")
    print("Have you registered to vote yet?")
else:
    print("Sorry, you are too young to vote.")
    print("Please register to vote as soon as you turn 18!")
print("Thank you and welcome again next time")
```

* The **if/else** block works when you only interested in two possible outcomes like the example above.

#### The if-elif-else Chain
* Often, youâ€™ll need to test more than two possible situations, and to evaluate these you can use Pythonâ€™s if-elif-else syntax. * Python executes only one block in an if-elif-else chain. 
* It runs each conditional test in order until one passes. 
* When a test passes, the code following that test is executed and Python skips the rest of the tests.
* Many real-world situations involve more than two possible conditions.

For example, consider an amusement park that charges different rates for
different age groups:
* Admission for anyone under age 4 is free.
* Admission for anyone between the ages of 4 and 18 is 5 dollars.
* Admission for anyone age 18 or older is 10 dollars.

```python
# Senior citizen discount = 5
age = int(input('Please enter your age: '))

if age <= 4:
    fee=0
elif age <= 18:
    fee=5
else:
    fee=10



    
print(f"Your admission cost is ${fee}")
```

- You can use as many elif blocks in your code as you like. eg in the example above the amusement park can introduce a discount for seniors, you can take care of that with another elif statement.
- Sometimes you can even skip the else block altogether.(Python doesn't require an else statement at the end of the if-elif chain.

### Assignment
* **BMI Calculator 2.0**

#### Instructions
* Write a program that interprets the Body Mass Index (BMI) based on a user's weight and height.
* It should tell the the interpretation of their BMI based on the BMI value.
    * Under 18.5 they are **underweight**.
    * Over 18.5 but below 25 they have **normal weight**.
    * Over 25 but below 30 they are **slightly overweight**.
    * Over 30 but below 35 they are **obese**.
    * Above 35 they are **clinically obese**.

### Solution

```python

weight = float(input("Enter yout weight in kilograms: "))
height = float(input("Enter your height in metres: "))

bmi = round(weight/(height **2), 0)

if bmi < 18.5:
    interpretation = "underweight"
elif bmi <= 25:
    interpretation = "normal weight"
elif bmi < 30:
    interpretation = "slightly overweight"
elif bmi < 35:
    interpretation = "obese"
else:
    interpretation = "clinically obese"
print("Your BMI is ", bmi)
print("You are ", interpretation)
```

### Use the instructions provided below to build a leap year calculator.
**Break down the problem and implement the solution in Python**

#### Instructions:
- A leap year is divisible by 4
- EXCEPT every year that is divisible by 100
- UNLESS the year is evenly divisible by 400

##############################################

e.g The year 2000
* 2000 / 4 = 500 (Leap)
* 2000 / 100 = 20 (Not leap)
* 2000 / 400 = 5 (Leap)
So 2000 is a leap year.

But 2100 is not a leap year because:

* 2100 / 4 = 525 (Leap) 
* 2100 / 100 = 21 (Not Leap)
* 2100 / 400 = 5.25 (Not leap)

#### Project 1 FLowchart

```python
Image(filename='leap_year_flowchart.png') 
```

### Solution

```python
year = int(input("Enter year: "))

if year % 4 != 0:
    leap = False
else:
    if year % 100 != 0:
        leap = True
    else:
        if year % 400 == 0:
            leap = True
        else:
            leap = False

if leap:
    print(f"{year} is a leap year")
else:
    print(f"{year} is NOT leap year")
```
